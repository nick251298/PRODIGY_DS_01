```python
# Importing necessary libraries
import matplotlib.pyplot as plt
import pandas as pd

# Loading data from CSV files
df = pd.read_csv(r"C:\Users\varun\Desktop\PRODIGY\Prodigy T1\world_population.csv")

# Create histograms for 2022, 2000, and 1990 population distributions
plt.figure(figsize=(15, 10))

# 2022 Population
plt.subplot(3, 1, 1)  # 3 rows, 1 column, first plot
plt.hist(df['2022 Population'], bins=30, color='skyblue', edgecolor='black')
plt.title('2022 Population Distribution Across Countries')
plt.xlabel('Population in 2022')
plt.ylabel('Frequency')

# 2000 Population
plt.subplot(3, 1, 2)  # 3 rows, 1 column, second plot
plt.hist(df['2000 Population'], bins=30, color='lightgreen', edgecolor='black')
plt.title('2000 Population Distribution Across Countries')
plt.xlabel('Population in 2000')
plt.ylabel('Frequency')

# 1990 Population
plt.subplot(3, 1, 3)  # 3 rows, 1 column, third plot
plt.hist(df['1990 Population'], bins=30, color='lightcoral', edgecolor='black')
plt.title('1990 Population Distribution Across Countries')
plt.xlabel('Population in 1990')
plt.ylabel('Frequency')

# Save the individual histograms as one figure with high DPI
plt.tight_layout()
plt.savefig('C:/Users/varun/Desktop/PRODIGY/population_histograms_comparison.png', dpi=300)

# Display the plot
plt.show()

# Close the plot to avoid display issues
plt.close()

# Create an overlapping histogram to compare 2022, 2000, and 1990 populations
plt.figure(figsize=(10, 6))

plt.hist(df['2022 Population'], bins=30, alpha=0.5, label='2022 Population', color='skyblue', edgecolor='black')
plt.hist(df['2000 Population'], bins=30, alpha=0.5, label='2000 Population', color='lightgreen', edgecolor='black')
plt.hist(df['1990 Population'], bins=30, alpha=0.5, label='1990 Population', color='lightcoral', edgecolor='black')

plt.title('Population Distribution Comparison (2022 vs 2000 vs 1990)')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.legend()

# Save the comparison plot with high DPI
plt.savefig('C:/Users/varun/Desktop/PRODIGY/population_distribution_comparison.png', dpi=300)

# Display the comparison plot
plt.show()

# Close the plot
plt.close()
```
